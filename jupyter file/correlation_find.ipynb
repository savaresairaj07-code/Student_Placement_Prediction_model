```python
numeric_col=df.select_dtypes('number').columns.to_list()
```

```python
catagorical_col=df.select_dtypes('object').columns.to_list()
```

```python
# df['Placement_enco']=df['Placement'].map({"No":0,"Yes":1})
placement_encoded = df['Placement'].map({'Yes':1,"No":0})
```

```python
# df['Intership_Ex_enco']=df['Internship_Experience'].map({"Yes":1,"No":0})
internship_encoded = df['Internship_Experience'].map({'Yes':1,"No":0})
```

```python
df['Placement']=placement_encoded
df['Internship_Experience']=internship_encoded
```

### Numeric_To_BInary

```python
numeric_correlation =pd.DataFrame()
```

```python
from scipy.stats import pointbiserialr
correlation=[]
p_value=[]
for col in numeric_col:
    corr, pval = pointbiserialr(df[col], df['Placement'])
    correlation.append(corr)
    p_value.append(pval)
    print(f"Correlation of {col}:", corr, "p-value:", pval)

numeric_correlation['features']= numeric_col
numeric_correlation['correlation'] = correlation
numeric_correlation['p-value'] = p_value
```

```python
numeric_correlation
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

table = pd.crosstab(df['Internship_Experience'], df['Placement'])
chi2, p, dof, ex = chi2_contingency(table)
print('f p < 0.05 => significant association between the two variables.')
print("Chi-square:", chi2, "p-value:", p)
```

```python
n = table.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(table.shape) - 1)))
print("Cramér’s V:", cramers_v)
```

```python
numeric_correlation['abs_correlation'] = np.abs(numeric_correlation['correlation'])
```
